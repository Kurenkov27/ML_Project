```python
import os
import pandas as pd
import numpy as np
import tensorflow.keras as keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../datasets/autoria/autoria_clean_data.csv')
orig_df = df.copy()
```

```python
df.head()
```

```python
# Selecting features and output
X = df[['mileage_kkm', 'year_made', 'engine_size']].head(400)
y = df['price_USD'].head(400)
```

```python
# DIviding data into training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = np.expand_dims(X_train, 1)
X_test = np.expand_dims(X_test, 1)
```

```python
regressior = Sequential()
regressior.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (X_train.shape[1], 3)))
regressior.add(Dropout(0.2))

regressior.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
regressior.add(Dropout(0.3))

regressior.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
regressior.add(Dropout(0.4))

regressior.add(LSTM(units = 120, activation = 'relu', input_shape = (X_train.shape[1], 3)))
regressior.add(Dropout(0.5))

regressior.add(Dense(units = 1))
```

```python
regressior.summary()
```

```python
regressior.compile(optimizer='adam', loss='mean_squared_error')
```

```python
regressior.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
y_pred = regressior.predict(X_test)
y_pred
```
